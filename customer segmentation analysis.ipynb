```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
K = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state=0)
    kmeans.fit(x)
    K.append(kmeans.inertia_)
plt.plot(range(1, 11), K)
plt.show()
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Assuming you have your data stored in the 'x' variable

# Instantiate a K-means model with the desired number of clusters
kmeans = KMeans(n_clusters=5)

# Fit the K-means model to your data
kmeans.fit(x)

# Obtain the cluster labels for each data point
y_KMeans = kmeans.predict(x)

# Plot the data points and cluster centers
plt.scatter(x[y_KMeans == 0, 0], x[y_KMeans == 0, 1], s=100, c='blue', label='Cluster 1')
plt.scatter(x[y_KMeans == 1, 0], x[y_KMeans == 1, 1], s=100, c='orange', label='Cluster 2')
plt.scatter(x[y_KMeans == 2, 0], x[y_KMeans == 2, 1], s=100, c='green', label='Cluster 3')
plt.scatter(x[y_KMeans == 3, 0], x[y_KMeans == 3, 1], s=100, c='purple', label='Cluster 4')
plt.scatter(x[y_KMeans == 4, 0], x[y_KMeans == 4, 1], s=100, c='red', label='Cluster 5')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=200, c='black', label='Centroids')
plt.title('K-means Clustering Algorithm')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.legend()
plt.show()
```
